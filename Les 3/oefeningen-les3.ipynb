### Oefening 2

```python
import numpy as np
from sympy.ntheory import isprime
import matplotlib.pyplot as plt

mersenneGetallen = [(2**x-1) for x in range(1,1001)]
mersennePriemgetallen = []

yArray = []

for mersenneGetal in mersenneGetallen:

    if isprime(mersenneGetal):

        if len(yArray)>0:
            yArray.append(yArray[-1]+1)
        else:
            yArray.append(1)

        mersennePriemgetallen.append(mersenneGetal)
    
    else:

        if len(yArray)>0:
            yArray.append(yArray[-1])
        else:
            yArray.append(0)

# Arange eindwaarde is niet inclusief
xx = np.arange(1,1001,1)
# Linspace eindwaarde is wel inclusief
xx = np.linspace(1,1000,1000)

print(mersennePriemgetallen)

plt.figure()
plt.step(xx,np.array(yArray))
plt.xlim(0,1000)
plt.show()
```

```python
import matplotlib.pyplot as plt
import sympy as sp

mersenneGetallen = [(2**x-1) for x in range(1,1001)]
mersennePriemgetallen = [n for n in range(1000) if sp.isprime(mersenneGetallen[n])]

print(mersennePriemgetallen)
print(range(len(mersennePriemgetallen)))

plt.figure()
plt.step(mersennePriemgetallen, range(len(mersennePriemgetallen)))
plt.xlim(0,1000)
plt.show()
```

```python
import sympy as sp

rechterlid = sp.Matrix([1,2,3])

stelsel = sp.Matrix([
    [1,2,4],
    [3,8,14],
    [2,6,11]
])

print("Determinant: ", stelsel.det())

# Oplossing met behulp van matrixrekenen
print("Oplossing 1: ", stelsel.solve(rechterlid))

opl = stelsel.inv()*rechterlid

# Oplossing met behulp van matrixrekenen
print("Oplossing 2: ", opl)
```

### Oefening 4

```python
import time

n = 100

A = np.array([
    [(i+j) for i in range(n)] for j in range(n)
])
B = np.array([
    [(i*j) for i in range(n)]for j in range(n)
])

tic = time.process_time()
C = np.dot(A,B)
toc = time.process_time()

print(toc-tic)

tic = time.process_time()
C2 = np.zeros((n,n))

for k in range(n):

    teSomeren = np.zeros((n,n))

    for i in range(n):

        for j in range(n):

            teSomeren[i,j] = A[i,k]*B[k,j]

    C2 += teSomeren
toc = time.process_time()

print(toc-tic)
```

### Oefening 5

```python
import sympy as sp

x,y = sp.symbols('x y')

def f(x,y): return x**3-x+y**3-y
def df(x0,y0): return (sp.diff(f(x,y),x).subs(x,x0), sp.diff(f(x,y),y).subs(y,y0))
def ddf(x0,y0): return (sp.diff(df(x,y),x).subs(x,x0), sp.diff(df(x,y),y).subs(y,y0))

sp.plotting.plot3d(f(x,y),(x,-2,2),(y,-2,2))

print(sp.solve(df(x,y),x,y))
```
