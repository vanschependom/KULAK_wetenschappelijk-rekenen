# Uitbreidingsoefeningen les 2

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')

def f(x):
    return x**3/(x**2-1)

def Df(x0):
    return sp.diff(f(x),x).subs(x,x0)

def D2f(x0):
    return sp.diff(f(x),x,2).subs(x,x0)
```

```python
f_num = sp.lambdify(x,f(x),"numpy")
Df_num = sp.lambdify(x,Df(x),"numpy")
D2f_num = sp.lambdify(x,D2f(x),"numpy")
```

```python
xx = np.linspace(-10,10,1000)
Y=f(xx)
Y[abs(Y)>10]=np.nan

plt.figure()


plt.subplot(3, 1, 1)

plt.plot(xx, Y, 'g', lw=2, label="f")
plt.plot([-100,100], [0,0], 'b', lw=1)
plt.plot([-100,100], [f(-np.sqrt(3)),f(-np.sqrt(3))], 'b', lw=1)
plt.plot([-100,100], [f(np.sqrt(3)),f(np.sqrt(3))], 'b', lw=1)
plt.axvline(x=-1, color='r', linestyle='--', lw=1)
plt.axvline(x=1, color='r', linestyle='--', lw=1)
plt.plot(xx, xx, 'r--', lw=1)
plt.xlim(-3,3)
plt.ylim(-8,8)
plt.scatter([-np.sqrt(3), 0, np.sqrt(3)],[f(-np.sqrt(3)),0,f(np.sqrt(3))],marker="o", c="r", zorder=2)
plt.legend()


plt.subplot(3, 1, 2)

plt.plot(xx, Df_num(xx), 'g', lw=2, label="f'")
plt.legend()
plt.xlim(-3,3)
plt.ylim(-5,5)
plt.plot([-10,10],[0,0],"k",lw=1)
plt.scatter([-np.sqrt(3), 0, np.sqrt(3)],[Df_num(-np.sqrt(3)),0,Df_num(np.sqrt(3))],marker="o", c="r", zorder=2)


plt.subplot(3, 1, 3)

Y2=D2f_num(xx)
Y2[abs(Y2)>100]=np.nan
plt.plot(xx, Y2, 'g', lw=2, label="f''")
plt.legend()
plt.xlim(-3,3)
plt.plot([-10,10],[0,0],"k",lw=1)
plt.ylim(-50,50)
plt.axvline(x=-1, color='r', linestyle='--', lw=1)
plt.axvline(x=1, color='r', linestyle='--', lw=1)
plt.scatter([0],[0],marker="o", c="r", zorder=2)

plt.show()
```
